```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("tips.csv")
```

```python
data
```

```python
sns.relplot(x="total_bill", y="tip",data=data)
```

Using Hue, which adds extra variable to the plot to classify
takes both categorical and nuemrical data 

```python
sns.relplot(x="total_bill", y="tip",hue="smoker", data=data)
```

Using style Parameter: takes categorical data, as well as numerical data

```python
sns.relplot(x="total_bill", y="tip",hue="smoker",style="sex",data=data)
```

using size parameter: Takes only numerical data 

```python
sns.relplot(x="total_bill", y="tip",hue="smoker",style="sex",size="size",data=data)
```

All these with the scatter plot 

```python
sns.scatterplot(x="total_bill", y="tip",hue="smoker",style="sex",size="size",data=data)
```

Categorical Data: on x-axis and numerical data on y-axis

```python
sns.catplot(x="day", y="tip", data=data) #kind is stripe (by default)
#you can add jitter(noise) as well, to see variations in the data plot
```

```python
sns.catplot(x="day", y="tip",jitter=2, data=data)
```

```python
#using kind=swarm
sns.catplot(x="day", y="tip", kind="swarm",data=data)
```

```python
#you can add hue as well

sns.catplot(x="day", y="tip",kind="swarm",hue="smoker" ,data=data)
```

Barplot and CountPlot 

```python
sns.catplot(x="day", y="total_bill", data=data, kind="bar")
```

```python
sns.barplot(x="smoker", y="tip", data=data, hue="sex")
```

HeatMaps : drawn between 2 categorical data, sometimes for numerical data but then they have to be discrete

```python
data
```

```python
data_pivot_table=data.pivot_table(values="tip", columns="smoker", index="sex", aggfunc=sum)

sns.heatmap(data_pivot_table)
```

```python
sns.heatmap(data_pivot_table, annot=True, linewidths=0.4, cmap="winter")
```

ClusterMaps: clusters based on the distance (ex: euclidean)

```python
sns.clustermap(data_pivot_table)
```

```python
#you can change metrics, z_score etc
sns.clustermap(data_pivot_table, annot=True)
```

Jointplot: performs bivariate analysis on numerical data

gives, (scatter/hexplot/kdeplot/Regplot) polt between 2 variables and individual histograms

```python
sns.jointplot(x="total_bill", y="tip", kind="scatter", data=data)
```

```python
sns.jointplot(x="total_bill", y="tip", kind="hex", data=data)
```

```python
sns.jointplot(x="total_bill", y="tip", kind="reg", data=data)
```

```python
sns.jointplot(x="total_bill", y="tip", kind="kde", data=data)
```

PairPlots: Very Helpful --> helps to differentiate which column is more userful 

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue="sex")
```

DistPlots --> Distribution Plots

performs univariate analysis on continous data

combines 3 grpahs: 1.histogram 2. Kernal Density Estimation (PDF-Probability Density Function) 3.Rugplot 

```python
data.head(2)
```

```python
sns.distplot(data['tip'])
```

```python
#only histogram
sns.distplot(data['tip'], kde=False)
```

```python
#only kde

sns.distplot(data['tip'], hist=False)
```

```python
#only Rug

sns.distplot(data['tip'], hist=False, kde=False)
```

```python
#
def smoker_numeric(x):
    if x=="No":
        return 0
    if x=="Yes":
        return 1

data['smoker_numeric'] = data['smoker'].apply(smoker_numeric)
```

```python
data
```

```python
#distplot of tips given by smoker and a nonsmoker comparision

sns.distplot(data[data['smoker_numeric']==0]['tip'])
sns.distplot(data[data['smoker_numeric']==1]['tip'])
```
